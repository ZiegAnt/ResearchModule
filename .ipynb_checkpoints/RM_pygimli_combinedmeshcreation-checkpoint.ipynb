```python
import pygimli as pg
import pygimli.meshtools as mt
```

```python
layer = pg.load("mesh_layermodel.poly")
layer_mesh = mt.createMesh(layer)
pg.show(layer_mesh)
```

```python
diatreme = pg.load("mesh_diatrememodel.poly")
```

```python
pg.show(diatreme, fillRegion=False)
```

```python
world = mt.createWorld([0,-70], [150, 0], layers=[-10,-30,-60])
pnts = [[20,-20],
        [30,-5],
        [50,-9],
        [70,-5],
        [80,-20]]
top = mt.createPolygon(pnts, addNodes=5, interpolate="spline")
left = mt.createLine([30,-70], pnts[0])
right = mt.createLine([70,-70], pnts[-1])
geom = world + top + left + right
pg.show(geom)
```

```python
for y in -9, -20, -50, -65:
    geom.addRegionMarker([50, y], 5)

for marker, y in enumerate([-20, -50, -65]):
    geom.addRegionMarker([100, y], marker + 2)

pg.show(geom)
```

```python
pg_mesh = mt.createMesh(geom, area=10)
pg.show(pg_mesh)
```
